```python
def unique_nodes(lines):
    all_raw = []
    for line in lines:
        a, b = line.split('-')
        all_raw.append(a)
        all_raw.append(b)
    
    uniques = np.unique(all_raw)
    return list(uniques)
```

```python
def get_nodes_map(lines, nodes):
    connections = [line.split('-') for line in lines]
    
    nodemap = {}
    for node in nodes:
        nodemap[node] = []
        for conn in connections:
            if node == conn[0]:
                nodemap[node].append(conn[1])
            elif node == conn[1]:
                nodemap[node].append(conn[0])
        
        if 'start' in nodemap[node]:
            nodemap[node].remove('start')
                
    return nodemap
```

```python
def option_used(queue_option, saved_paths):
    '''check if potential path is used an endpath'''
    
    # for path in saved_paths:
    #     if set(queue_option).issubset(set(path)):
    #         return True
    
    howfar = len(queue_option)
    for path in saved_paths:
        if queue_option == path[:howfar]:
            return True
    
    return False
```

```python
# test 
to_check = ['start', 'A', 'c', 'A', 'b'] + ['A']
saved = [['start', 'A', 'c', 'A', 'b', 'A', 'end']]
assert option_used(to_check, saved) == True

to_check = ['start', 'A', 'c', 'A', 'b'] + ['end']
saved = [['start', 'A', 'c', 'A', 'b', 'A', 'end']]
assert option_used(to_check, saved) == False
```

```python
def looseEnd(queue, options):
    
    all_low = [option.islower() for option in options]
    
    # capital in there, so an option
    if all(all_low) != True:
        return False
    
    all_taken = [option in queue for option in options]
        
    if all(all_taken) == True:
        return True
    else:
        return False
```

```python
testqueue2 = ['start', 'A', 'c', 'A', 'b']
testoption = ['b']
looseEnd(testqueue2, testoption)

testqueue3 = ['start', 'HN', 'dc']
testoption3 = ['dc']
assert looseEnd(testqueue3, testoption3) == True
```

```python
# start ['A', 'b']
queue = ['start']
saved_paths = []

testrounds = 0

while len(queue) != 0:
    # iterates per queue position check
    
        
    if testrounds == 30:
        break
    
    done = 0
    current_node = queue[-1]
    options = nodes_map[current_node]
    
    for k, option in enumerate(options):
        print(testrounds, queue, current_node, '-option-',option)
        
        # lowercase already used
        if (option.islower()) & (option in queue):
            option = None
            continue
                    
        # if end option already used 
        if option == 'end' and (queue + ['end'] in saved_paths):
            option = None
            continue
        
        # if loose end (all option lowers in queue already)
        if option.islower() & (option != 'end'):
            if looseEnd(queue, nodes_map[option]):
                print('loose end',queue, option)
                option = None
                continue
        
        # if next option in saved path skip
        if option_used(queue + [option], saved_paths):
            option = None
            continue
        
        queue.append(option) #
        break
        
        done = True
    
    # if no option was appended(option = NOne)
    if option == None:
        queue.pop()
        continue
        
    if (queue + ['end'] in saved_paths):
        queue.pop()
        testrounds += 1
        continue
    
    # if end & end option not in saved paths
    if option == 'end':
        saved_paths.append(queue.copy())
        print('saving, popping', queue)
        queue.pop()
        
        #break
        
        
    testrounds += 1
```
